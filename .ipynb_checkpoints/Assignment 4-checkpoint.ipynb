# SU - Assignment 4: Decision Trees

```python
import pandas as pd
import numpy as np
import re
from sklearn.model_selection import KFold, cross_val_score, train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import plot_tree
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Read and view the dataset
df = pd.read_csv("data/House_Price.csv")
df.head(10)
```

### Preprocessing

```python
#Create dummies from discrete parameters and turn booleans to 1/0 
df = pd.get_dummies(df,columns=['waterbody'])
df[['airport','bus_ter']] = df[['airport','bus_ter']].applymap(lambda x: 1 if x=='YES' else 0)
df = df.dropna()
```

```python
df
```

```python
X = df.drop('price',axis=1)
y = df['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=141)
```

```python
#min max scale
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

### Sklearn - DecisionTreeRegressor

```python
baseline_tree = DecisionTreeRegressor(criterion='mse',min_samples_leaf=10)
```

```python
k = KFold(n_splits=5,shuffle=True)
scores = cross_val_score(baseline_tree,X_train,y_train,cv=k,scoring='neg_mean_squared_error')
```

```python
scores
```

```python
print(f"Average RMSE using 5-fold Cross-Validation on scaled data: {(-scores.mean())**0.5}")
```

```python
baseline_tree.fit(X_train,y_train)
y_pred = baseline_tree.predict(X_test)
rmse_test = mean_squared_error(y_test,y_pred,squared=False)
```

```python
print(f"RMSE on the test set after scaling: {rmse_test}")
```

```python
# Scatterplot of predictions vs. ground truth
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.title('Actual Prices vs Predicted')
plt.plot(np.linspace(min(y_test), max(y_test), 100), np.linspace(min(y_test), max(y_test), 100), color='red', linestyle='--',label="Expected performance for perfect regressor")
plt.legend()
plt.show()
```

### Custom tree regressor

```python
class MyTreeRegressor:
    def __init__(self,max_depth=None,min_samples_leaf=1):
        self.max_depth = max_depth
        self.min_samples_leaf = min_samples_leaf
        self.tree = None
        
    def fit(self,X,y):
        self.tree = self._build(X,y,depth=0)
        
    def predict(self,X):
        assert self.tree is not None, "Tree is not fitted!"
        return np.array([self._traverse(x,self.tree) for x in X])
        
    def _build(self,X,y,depth):
        num_samples, num_features = X.shape
        if depth == self.max_depth or num_samples <= self.min_samples_leaf or len(np.unique(y)) == 1:
            return np.mean(y)
        
        # Find the best split feature and value based on information gain
        best_split_feature, best_split_value = self._find_best_split(X,y)
        
        if best_split_feature is not None:
            #Split based on best feature and value, then build two subtrees
            left_idxs = X[:,best_split_feature] <= best_split_value
            right_idxs = X[:,best_split_feature] > best_split_value
            
            left_subtree = self._build(X[left_idxs],y[left_idxs],depth+1)
            right_subtree = self._build(X[right_idxs],y[right_idxs],depth+1)
            return {
                'feature_index': best_split_feature,
                'split_value': best_split_value,
                'left': left_subtree,
                'right': right_subtree,
            }
        else:
            return np.mean(y)
            
        
    
    def _traverse(self,x,node):
        if isinstance (node, dict):
            feature_index = node['feature_index']
            split_value = node['split_value']
            
            if x[feature_index] <= split_value:
                return self._traverse(x,node['left'])
            else:
                return self._traverse(x,node['right'])
        else: #is leaf, return value
            return node
    
    def _find_best_split(self,X,y):
        #We are brute forcing through all combinations...
        num_samples, num_features = X.shape
        best_split_feature = None
        best_split_value = None
        best_information_gain = -np.inf

        #Each value for each feature is tried as a split point... 
        for feature in range(num_features):
            fvalues = np.unique(X[:,feature])
            for value in fvalues:
                left_idxs = X[:,feature] <= value
                right_idxs = X[:,feature] > value
                left_y = y[left_idxs]
                right_y = y[right_idxs]
                
                #...and the combination with the best information gain is chosen
                if len(left_y) > 0 and len(right_y) > 0:
                    information_gain = self._information_gain(left_y, right_y)
                    if information_gain > best_information_gain:
                        best_information_gain = information_gain
                        best_split_feature = feature
                        best_split_value = value
                        
        return best_split_feature,best_split_value
                    
    def _information_gain(self,left_y,right_y):
        parent_entropy = self._entropy(np.concatenate((left_y,right_y)))
        num_left = len(left_y)
        num_right = len(right_y)
        total_samples = num_left + num_right
        left_weight = num_left/total_samples
        right_weight = num_right/total_samples
        
        left_entropy = self._entropy(left_y)
        right_entropy = self._entropy(right_y)
        
        return parent_entropy - (left_weight * left_entropy) - (right_weight * right_entropy)
    
    def _entropy(self,y):
        #Shannon's entropy
        unique_classes, class_counts = np.unique(y, return_counts=True)
        P = class_counts / len(y)
        return - np.sum(P * np.log2(P + 1e-10)) #To avoid division by zero
        
```

```python
my_tree = MyTreeRegressor(max_depth=10, min_samples_leaf=10)
```

```python
my_tree.fit(X_train,y_train)
y_pred = my_tree.predict(X_test)
rmse_test = mean_squared_error(y_test,y_pred,squared=False)
```
